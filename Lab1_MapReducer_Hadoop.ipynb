#4   Manipulation du MapReduce avec Python



````python
import os
os.environ['PATH'] += ':/content/hadoop-3.3.5/bin'
os.environ['PATH'] += ':/content/hadoop-3.3.5'
````

Charger les deux fichiers jobmap et jobred qui correspendsent respectivement au mapper et au reducer du job wordcount

````python
#create mapper and reducuer
!sudo chmod 777 /content/jobmap.py
!sudo chmod 777 /content/jobred.py
!sudo chmod  -R  777 /content
````

Executer le job wordcount en se servant du code ci-dessous:

````python
#Hadoop streaming is a utility that comes with the Hadoop distribution.
#The utility allows you to create and run Map/Reduce jobs with any executable
#or script as the mapper and/or the reducer. For example:


!mapred streaming \
  -input hdfs://localhost:9000/data/wordcount/input/speech.txt \
  -output hdfs://localhost:9000/data/wordcount/output/6 \
  -mapper  'python jobmap.py' \
  -reducer  'python jobred.py' \
  -file /content/jobmap.py \
  -file /content/jobred.py


````

````python
!sudo /content/hadoop-3.3.5/bin/hdfs dfs -ls /data/wordcount/output/6
````

````python
!sudo /content/hadoop-3.3.5/bin/hdfs dfs -cat /data/wordcount/output/6/part-00000
````
